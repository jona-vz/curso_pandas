```python
#***2.6.1 Creating index ***
import pandas as pd
cast = pd.read_csv('cast.csv', index_col=None)
cast.head()
```

```python
%%time

#data acces without indexing
cast[cast['title']== 'Macbeth']
```

```python
%%timeit #for more precise results as it run the shell various times and display the average time 

#data acces without indexing
cast[cast['title']== 'Macbeth']
```

```python
#below line will not working for multiple index
# c =  cast.set_index('title') so instead we use:
c = cast.set_index(['title'])
c.head()
```

```python
%%time

#data access with indexing
#note that there is minor performance improvement
c.loc['Macbeth']
```

```python
%%timeit

#data access with indexing
#note that there is minor performance improvement
c.loc['Macbeth']
```

```python
cs =  cast.set_index(['title']).sort_index()
cs.tail(4)
```

```python
%%time

# data access with indexing
# note that there is huge performance improvement
cs.loc['Macbeth']
```

```python
%%timeit

# data access with indexing
# note that there huge performance improvement
cs.loc['Macbeth']
```

```python
#***2.6.2 Multiple_index***
#data with two index i.e. title and n
cm = cast.set_index(['title', 'n']).sort_index()
cm.tail(30)
```

```python
cm.loc['Macbeth']
```

```python
#show Macbeth with rankinkg 4-18
cm.loc['Macbeth'].loc[4:18]
```

```python
#show Macbeth with rankinkg 4
cm.loc['Macbeth'].loc[4]
```

```python
#***2.6.3 Reset index***
cm.head(2)
```

```python
#remove 'n' from index
cm = cm.reset_index('n')
cm.head(2)
```

```python
cm = cm.reset_index(level='title')
cm.head(2)
```
